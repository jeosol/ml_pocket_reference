## Imputing Data

- The general task of defining values for missing values is called imputation
- If you are imputing data, you will need to build a pipeline and use the same imputation logic during model creation and prediction time.
- The SimpleImputer class in scikit-learn will handle mean, median and most frequent feature values

```python
from sklearn.impute import SimpleImputer
num_cols = df.select_dtypes(include="number").columns
```

```python
num_cols
```

```python
im = SimpleImputer() # uses mean by default
imputed = im.fit_transform(df[num_cols])
```

- provide stategy='median' or strategy='most_frequent' to change the replace value to median or most common 
respectively
- To use a constant value, use strategy='constant' and with fill_value=-1
- frequent or constant strategies can be used with numeric or string data
- mean and median strategies require numeric data

## Adding Indictor Columns

- The lack of data may provide some signal to a model
- THe pandas library can add a new column to indicate that a value was missing


```python
# This function is used to convert a column to a indicator variable
# First to True/False via .isna() and the 1/0 via .astype()
def add_indicator(col):
    def wrapper(df):
        return df[col].isna().astype(int)
    
    return wrapper
```

```python
# This function is used to assign a new column with name 'cabin_missing' and returns all original columns in 
# addition to new ones. Existing columns that are assigned will be overwritted
df1 = df.assign(cabin_missing=add_indicator("cabin"))
```
